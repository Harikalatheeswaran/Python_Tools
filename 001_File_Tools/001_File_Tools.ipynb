````python
from rich.prompt import Prompt as p
from rich.console import Console
import base64 as b
from pathlib import Path 

console = Console()

file_path = Path(r"C:\Users\harik\Desktop\PythonCodes\003_Tools\001_File_Tools\dummy.txt")

# ________________________________________________________________________________________________________________________________________________

def encode_base64(text : str, cycles : int = 1) -> str:
    """
    Encodes a string in base64, optionally multiple times.

    Args:
        text (str): Input string to encode.
        cycles (int): Number of encoding cycles (default 1). 0 or negative returns original.

    Returns:
        str: Base64-encoded string (multi-encoded if cycles > 1).
    """
    if not isinstance(text, str):
        raise ValueError("Input 'text' must be a string.")
    
    temp = text
    cycles = max(0, cycles)  # Clamp to non-negative

    if cycles > 9:
        warning = f"ðŸš¨ Whoa, base64 Gremlin alert! ðŸš¨\n>9 cycles? Your string will bloat ~33% *each* roundâ€”like a bad diet on steroidsâ€”turning 'hello' into a RAM-gobbling novel by cycle 10. Dialing back to 9: fun obfuscation without the felony-sized file!"
        #print("Whoa, you must be insane!\nText will become excruciatingly large.\nHence encoding only once!")
        print(warning)
        encoded_string = b.b64encode(bytes(temp, encoding="utf-8")).decode(encoding="utf-8")
        return encoded_string
    
    encoded_string = text  # Initialize to original in case cycles == 0
    
    for _ in range(cycles):
        encoded_string = b.b64encode(bytes(temp, encoding="utf-8")).decode(encoding="utf-8")
        temp = encoded_string
    
    return encoded_string   

# ________________________________________________________________________________________________________________________________________________

def decode_base64(text : str, cycles : int = 1) -> str:
    """
    Decodes a base64 string, optionally multiple times.

    Args:
        text (str): Input base64 string to decode.
        cycles (int): Number of decoding cycles (default 1). 0 or negative returns original.

    Returns:
        str: Base64-decoded string (multi-decoded if cycles > 1).

    Note : 
    binascii.Error is Python's way of yelling "Hey, that binary-to-ASCII conversion went sideways!" 
    It's thrown by the binascii module (which powers stuff like base64.b64decode) when the input is garbage for the jobâ€”like feeding it non-base64 chars, 
    wrong padding (base64 needs multiples of 4 chars with '=' pads), or just plain invalid bytes. 
    """
    if not isinstance(text, str):
        raise ValueError("Input 'text' must be a string.")
    
    temp = text
    cycles = max(0, cycles)  # Clamp to non-negative

    if cycles > 9:
        Warning = f"ðŸš¨ Whoa, base64 Onion alert! ðŸš¨\n>9 decodes? It'll shrink your string fine at first, but then hit padding pitfallsâ€”like Russian roulette with '=' signsâ€”unraveling into 'invalid base64' gibberish by round 10. Capping at 9: safe unraveling without the error apocalypse!"
        #print("Whoa, you must be insane!\nMulti-decoding could unravel into gibberish.\nHence decoding only once!")
        print(Warning)
        try:
            decoded_string = b.b64decode(bytes(temp, encoding="utf-8")).decode(encoding="utf-8")
            return decoded_string
        except (b.binascii.Error, UnicodeDecodeError):
            print("Warning: Invalid base64 inputâ€”returning original.")
            return text
    
    decoded_string = text  # Initialize to original in case cycles == 0
    
    for i in range(cycles):
        try:
            decoded_string = b.b64decode(bytes(temp, encoding="utf-8")).decode(encoding="utf-8")
            temp = decoded_string
        except (b.binascii.Error, UnicodeDecodeError):
            print(f"Warning: Decode failed at cycle {i+1}.\nReturning partial result.")
            return decoded_string  # Return what we have so far
    
    return decoded_string 

# ________________________________________________________________________________________________________________________________________________

def encode_file_base64(file_path : Path):
    """_summary_

    Args:
        file_path : Path of the file
    """
    with open(file_path, "r") as fh:
        data = fh.readlines()
        #print(data)

    encoded_file_path = f"encoded_{file_path.parts[-1]}"

    with open(encoded_file_path, "w") as fh:
        for line in data:
            l = b.b64encode(bytes(line, encoding="utf-8")).decode(encoding="utf-8")
            fh.write(f"{l}\n")
            
        print(f"[+]{file_path.parts[-1]} is encoded with Base64 & is written to {encoded_file_path}")

# ________________________________________________________________________________________________________________________________________________

def decode_file_base64(file_path : Path):
    """_summary_

    Args:
        file_path : Path of the file
    """
    with open(file_path, "r") as fh:
        data = fh.readlines()
        #print(data)

    decoded_file_path = f"decoded_{file_path.parts[-1]}"

    with open(decoded_file_path, "w") as fh:
        for line in data:
            l = b.b64decode(bytes(line, encoding="utf-8")).decode(encoding="utf-8")
            fh.write(f"{l}")
            
        print(f"[+]{file_path.parts[-1]} is decoded with Base64 & is written to {decoded_file_path}")

# ________________________________________________________________________________________________________________________________________________

````
